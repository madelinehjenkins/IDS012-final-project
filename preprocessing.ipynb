```python
import numpy as np
import pandas as pd
import json
```

```python
vote_data_year = {}
vote_data_by_year = json.loads(open("senatevotedata.json").read())
#vote_data = pd.read_
for year in range(1989, 2014):
    vote_data_year[year] = list(pd.DataFrame(vote_data_by_year["year"][str(year)])['votes'])
    vote_data_year[year] = [bill['Yea'] for bill in vote_data_year[year]]
    #list of lists, each list is group of people who voted for the same bill
    df = pd.Series()
    bill_number = 0
    for bill in vote_data_year[year]:
        for person in bill:
            df = pd.concat([df,pd.Series(person).append(pd.Series([bill_number], index = ['bill_number']))], axis = 1)
        bill_number += 1
    df = df.transpose()
    vote_yes[year] = df    
    vote_yes[year] = vote_yes[year].dropna()

```

```python
for year in vote_yes:
    vote_yes[year] = vote_yes[year].reset_index()
    vote_yes[year] = vote_yes[year].drop('index', 1)
    
```

```python
for year in vote_yes: 
    vote_yes[year].to_csv(str(year) + '_yes_votes.csv')
```
